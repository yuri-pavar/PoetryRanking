# Utils

## rhymetagger

```python
def get_rhymes(data, rt_model):
  lst = []
  for txt in tqdm(data):
    txt = [i for i in txt.split('\n') if i != ''] # делим на строки и удаляем пустые
    output = rt_model.tag(txt, output_format=3)
    output = [i if i else -1 for i in output] # -1 вместо None для строк без рифм
    lst.append(output)
  return lst
```

## numblilbug - 3 class


```python
def get_3class(data, model, tokenizer):
  meanf_list, grammar_list, emotion_list = [], [], []
  for txt in tqdm(data):
    txt_tokenized = tokenizer(txt, return_tensors="pt",padding=True, truncation=True, max_length=50, add_special_tokens=True) # params от создателей
    txt_tokenized = txt_tokenized.to(DEVICE)
    with torch.no_grad():
      outputs = model(**txt_tokenized)
      meanf, grammar, emotion = (outputs.logits > 0.6).squeeze().int().cpu().numpy().tolist() # 0.6 порог от создателей
      meanf_list.append(meanf)
      grammar_list.append(grammar)
      emotion_list.append(emotion)

  return meanf_list, grammar_list, emotion_list
```

```python
pd.DataFrame(
    {
        'meanf': meanf_list,
        'grammar': grammar_list,
        'emotion': emotion_list,
        't': [1]*len(emotion_list)
        }
).groupby(['meanf', 'grammar', 'emotion'], as_index=False).t.count().sort_values(['meanf', 'grammar', 'emotion'])
```

```python
def get_syllable_stress_list(word, syllables):
    syllable_list = syllables#.split('|')
    stress_position = word.find("'")
    if stress_position == -1:
      return []
    current_position = 0
    stress_syllable_index = 0
    for i, syllable in enumerate(syllable_list):
        current_position += len(syllable)
        if current_position >= stress_position:
            stress_syllable_index = i
            break
    stress_list = [0] * len(syllable_list)
    stress_list[stress_syllable_index] = 1

    return stress_list


def get_stress(txt):
  return stress_rnn.put_stress(txt, stress_symbol="'", accuracy_threshold=0.75, replace_similar_symbols=True)


def get_syll(txt):
  return rusyll.token_to_syllables(txt)
```

```python
# 10 - horey
# 01 - yamb
# 100 - daktil
# 010 - amfibrakh
# 001 - anapest
#
def get_liter_size(poem):
  '''
    Функция возвращает схемы ударных и безударных слогов слов для каждой строки.
    Каждый самый глубокий список - это схема одного слова.
  '''
  poem_lst = []
  for line in poem.split('\n'):
    if line != '':
      line_lst = []
      for i in line.split():
        syll = rusyll.token_to_syllables(i)
        if len(syll) > 1: # только состоящие из 2 слогов и более
          w_list = get_syllable_stress_list(get_stress(i), get_syll(i))
          if len(w_list) > 0: # не берем те, где не проставилось ударение
            line_lst.append(w_list)
      poem_lst.append(line_lst)
  return poem_lst

def get_liter_size_coded(poem):
  '''
    Функция копирует get_liter_size за исключением выхода.
    Выход функции - закодированные схемы ударных и безударных слогов по правилу:
    результат = кол-во слогов + позиция 1/кол-во слогов
  '''
  poem_lst = []
  for line in poem.split('\n'):
    if line != '':
      line_lst = []
      for i in line.split():
        syll = rusyll.token_to_syllables(i)
        if len(syll) > 1: # только состоящие из 2 слогов и более
          w_list = get_syllable_stress_list(get_stress(i), get_syll(i))
          if len(w_list) > 0: # не берем те, где не проставилось ударение
            poem_lst.append(len(w_list) + (w_list.index(1)+1)/len(w_list))
  return poem_lst
```
